### Data Augmentation

Data Augmentation is a way of applying different types of transformation techniques on actual images, thus producing copies of the same image with alterations. This helps to train deep learning models on more image variations than what is present in the actual dataset.

This is especially useful when we do not have enough images to train the model on, hence we can use augmented images to enlarge the training set and provide more images to the model.

In order to perform data augmentation on images, we will use `ImageDataGenerator`

The `ImageDataGenerator` class in Keras is used for preprocessing and data augmentation of image data.

It generates batches of tensor image data with real-time data augmentation. 

The data is either directly loaded from memory or from disk. 

The data augmentation techniques include random transformations, such as flipping, rotating, scaling, and shifting of the images, to increase the diversity of the training data. 

This helps to prevent overfitting and improves the performance of the model on unseen data.

```python
# Apply transformation

from keras.preprocessing.image import ImageDataGenerator, img_to_array, image
import warnings
warnings.filterwarnings('ignore')


batch_size=60
# ImageDataGenerator() accepts the original data, randomly transforms it, and returns only the new, transformed data.
transformed_train_set = ImageDataGenerator( 
    rescale=1./255, # so that pixel value are in same range for all range.
    
    rotation_range=40, # Degree range for random rotations.
    
    width_shift_range=0.1,  # randomly shift images horizontally (fraction of total width.)
    
    height_shift_range=0.1,  # randomly shift images vertically (fraction of total height.)
    
    horizontal_flip=True, # whether to randomly flip images horizontally. 
    
    featurewise_std_normalization=True, # Divide inputs by std of the dataset, feature-wise.
    
    samplewise_std_normalization=True) # Divide each input by its std.

transform_validation_set = ImageDataGenerator(
    rescale=1./255,
    rotation_range=40,
    width_shift_range=0.1,  # randomly shift images horizontally
    height_shift_range=0.1,  # randomly shift images vertically
    horizontal_flip=True,
    featurewise_std_normalization=True,
    samplewise_std_normalization=True)

transform_test_set = ImageDataGenerator(
    featurewise_std_normalization=True,
    samplewise_std_normalization=True,
    rescale=1./255)
```

```python
# flow_from_directory() is a method in the ImageDataGenerator class in Keras
# it generates batches of augmented/preprocessed data from a directory of image files. 
# This method is used to generate data for training a deep learning model using a large number of image files.

# The method takes in several parameters that allow you to specify the configuration of the data being generated, including:

# flow_from_directory() method takes a path of a directory and generates batches of augmented data.


train_generator = transformed_train_set.flow_from_directory(
        train_set, # path to the directory that contains the image files.
    
        target_size=(224, 224), # the size to which all images found will be resized.
    
        batch_size=batch_size, # number of samples per batch.
    
        class_mode='categorical') # Determines the type of label arrays that are returned.
        # one of "categorical", "binary", "sparse", "input", or None.

validation_generator = transform_validation_set.flow_from_directory(
        validation_set,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical')

test_generator = transform_test_set.flow_from_directory(
        test_set,
        target_size=(224, 224),
        batch_size=batch_size,
        class_mode='categorical')
```

### MobileNetV2

The MobileNetV2 model is a lightweight and efficient convolutional neural network (CNN) architecture designed specifically for mobile and embedded vision applications.

```python
from keras.layers import Dense, Input, Dropout
from keras.applications.mobilenet_v2 import MobileNetV2
from keras.models import Model

# Input() method is used instantiate a Keras tensor. 
# A Keras tensor is a symbolic variable, which serves as the placeholder for an input to a model. 

input_tensor = Input(shape=(224, 224, 3)) # typically used to define the input layer(s) of a model in Keras, 
                                            # where the shape and dtype of the input tensors are specified.
    
"""Once the input tensor has been defined using the Input method,
    it can be used as the first argument in subsequent calls
    to other layer methods to build the model."""
    
model1 = MobileNetV2(
    include_top=False, # parameter determines whether the fully connected layer at the top of the network
                            # should be included or not.
    
    # When include_top=False, the network outputs intermediate activations 
    #(i.e., activations from the layer prior to the final dense layer),
    # and the fully connected layer at the top of the network is excluded.
    
    weights='imagenet', # pre-training on ImageNet
    
    input_tensor=input_tensor, # to use as image input for the model.
    
    input_shape=(224, 224, 3), # would like to use a model with an input img resolution that is (224, 224, 3)
    
    pooling='avg') # pooling mode for feature extraction as we have set include_top=False.

# The MobileNetV2() method returns a Keras model instance.

for layer in model1.layers:
    layer.trainable = True 
"""layer.trainable=True means that the layer's weights will be updated during training.
In other words, the backpropagation algorithm will adjust the weights in the layer in response 
to the loss during training. If layer.trainable=False, the weights in the layer will remain fixed
and will not be updated during training."""
```

Here, the `input tensor` is defined, and then the `dense layer`, `dropout layer`, and `output layer` are added to the model in succession. The resulting model can be compiled using the `compile` method, and then trained using the `fit` method.

`optimizer`: the optimizer to use when updating the model weights during training. The optimizer can be one of the built-in optimizers in Keras (e.g., `Adam`, `SGD`, `RMSprop`), or a custom optimizer created using the optimizers module.

`loss`: the loss function to use when evaluating the model during training. The loss function should be a string (for one of the built-in loss functions in Keras) or a callable that takes the true labels and predicted labels as inputs.

`metrics`: a list of metrics to use when evaluating the model during training and testing. The metrics can be strings (for one of the built-in metrics in Keras) or callables that take the true labels and predicted labels as inputs.

Here, the model is compiled using the `Adam` optimizer, the `categorical cross-entropy` loss function, and the `accuracy` metric. The resulting model can then be trained using the fit method.

```python
from keras.callbacks import ModelCheckpoint, ReduceLROnPlateau, EarlyStopping

train_length = len(train_generator.filenames)
validation_length = len(validation_generator.filenames)
test_length = len(train_generator.filenames)


# set a pointer for When to save the model weights
checkpoint = ModelCheckpoint(filepath='model_best_weights.hdf5', verbose=1, 
                               save_best_only=True)
# The verbose parameter is set to 1, so a progress bar-style output will be displayed during training.

# Reduce learning rate when loss doesn't improve after 5 epochs
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.1,
                              patience=5, min_lr=1e-8, verbose=1)

# Stop early if model doesn't improve after 12 epochs
early_stopping = EarlyStopping(monitor='val_loss', patience=12,
                              verbose=0, restore_best_weights=True)
# The restore_best_weights parameter takes a boolean value, and if it is set to True,
#the best weights found during training will be restored when the training is stopped early. 
```

```python
%%time
from keras.optimizers import Adam
import keras

activation_functions = ['relu', 'sigmoid', 'tanh'] # list of activation functions to test

test=0.0

for activation in activation_functions:
    
    print(f"Activation function: {activation}")

    dense_layer = Dense(512, activation=activation)(model1.output) # used to instantiate a dense layer in a neural network
    # returns a dense layer instance that can be added to a model
    # The output of the dense layer can then be passed to other layer instances to build up the model.

    """A dense layer is a fully connected layer, where each neuron in the layer is
    connected to every neuron in the previous layer. The dense layer can be used to
    model a wide range of functions, from simple linear regression to complex non-linear models."""

    # number of neurons in the dense layer is set 512, and these gone learn new weights
    # activation function to use for the layer is set relu

    dropout_layer = Dropout(.8)(dense_layer) #  drop out 80% of the neurons in the layer during each training step.
    # returns a dropout layer instance that can be added to a model
    # During training, the dropout layer will randomly drop out some of the neurons in the dense layer,
    # helping to prevent overfitting.

    predictions = Dense(211, activation='softmax')(dropout_layer)
    # When making predictions in the model, the dropout layer is turned off, and all neurons are used.


    # Model class in Keras is used to define a model for training, evaluation, and prediction. 
    # A model is a directed acyclic graph of layers that can be trained end-to-end to minimize a loss function.

    model = Model(inputs=model1.input, outputs=predictions)
    # inputs is input layer or tensor of model
    # outputs is output layer or tensor of model

    model.compile(optimizer=Adam(lr=0.0001), loss='categorical_crossentropy',
                 metrics=['accuracy'])
    # Train the model
    history = model.fit_generator(train_generator,
                        steps_per_epoch=train_length//batch_size,
                        epochs=100,
                        verbose=1,
                        callbacks=[checkpoint, reduce_lr, early_stopping],
                        validation_data=validation_generator,
                        validation_steps=validation_length//batch_size)
    
    test_accuracy = model.evaluate_generator(test_generator, steps=test_length//1, verbose=1)
    print(f"\nTest accuracy using {activation} : {test_accuracy[1]}\n")
    
    if test_accuracy[1]>test:
        test=test_accuracy[1]
        model.save('E:/sem8/models/cnn_model_'+activation+'.h5')
    else:
        pass
    
```
