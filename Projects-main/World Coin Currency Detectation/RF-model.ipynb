```python
import numpy as np 
import matplotlib.pyplot as plt
import glob
import cv2
import os
import seaborn as sns
import pandas as pd
from skimage.filters import sobel
```

```python
print(os.listdir("coins/data/"))
```

```python
#Resize images to
SIZE = 128
```

```python
# training data 

#Capture images and labels into arrays.
#Start by creating empty lists.
train_images = []
train_labels = [] 
for directory_path in glob.glob("coins/data/train/*"):
    label = directory_path.split("\\")[-1]
    print(label)
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg")):
        print(img_path)
        img = cv2.imread(img_path, cv2.IMREAD_COLOR) #Reading color images
        img = cv2.resize(img, (SIZE, SIZE)) #Resize images
        #img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) #Optional step. Change BGR to RGB
        train_images.append(img)
        train_labels.append(label)
```

```python
train_images = np.array(train_images)
train_labels = np.array(train_labels)
```

```python
train_images.shape
```

```python
train_labels.shape
```

```python
test_images = []
test_labels = [] 
for directory_path in glob.glob("coins/data/test/*"):
    fruit_label = directory_path.split("\\")[-1]
    for img_path in glob.glob(os.path.join(directory_path, "*.jpg")):
        img = cv2.imread(img_path, cv2.IMREAD_COLOR)
        img = cv2.resize(img, (SIZE, SIZE))
        #img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) #Optional
        test_images.append(img)
        test_labels.append(fruit_label)
```

```python
test_images = np.array(test_images)
test_labels = np.array(test_labels)
```

```python
test_images.shape
```

```python
test_labels.shape
```

```python
#Encode labels from text (folder names) to integers.
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
le.fit(test_labels)
test_labels_encoded = le.transform(test_labels)
le.fit(train_labels)
train_labels_encoded = le.transform(train_labels)
```

```python
#Split data into test and train datasets (already split but assigning to meaningful convention)
#If you only have one dataset then split here
x_train, y_train, x_test, y_test = train_images, train_labels_encoded, test_images, test_labels_encoded

# Normalize pixel values to between 0 and 1
x_train, x_test = x_train / 255.0, x_test / 255.0
```

```python
dtype=np.float64
x_train_loaded= np.memmap('E:/sem8/svm/X_train_scaled_data.dat.npy', dtype=dtype, mode='r',shape=(6413, 245760))
```

```python
%%time

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
import numpy as np

# Define the number of classifiers to use
n_classifiers = 10

# Define the size of each feature subset
subset_size = x_train_loaded.shape[1] // n_classifiers

# Train each classifier on a different feature subset
classifiers = []
for i in range(n_classifiers):
    # Determine the feature indices for this subset
    start = i * subset_size
    end = (i + 1) * subset_size
    subset_indices = np.arange(start, end) % x_train_loaded.shape[1]

    # Create a new classifier and fit it to the subset of features
    classifier = RandomForestClassifier()
    classifier.fit(x_train_loaded[:, subset_indices], y_train)

    # Add the classifier to the list of classifiers
    classifiers.append(classifier)
```

```python
x_test_loaded=np.memmap('E:/sem8/svm/x_test_scaled_data.dat.npy', dtype=dtype, mode='r', shape=(844, 245760))
x_test_loaded
```

```python
x_test_loaded.shape
```

```python
# Use the classifiers to make predictions on the test set
y_pred = np.zeros((x_test_loaded.shape[0], np.max(y_train) + 1))
for classifier in classifiers:
    y_pred += classifier.predict_proba(x_test_loaded[:, subset_indices])
y_pred /= n_classifiers
```

```python
y_pred
```

```python
# Evaluate the performance of the ensemble classifier
accuracy = np.mean(y_test == np.argmax(y_pred, axis=1))
print("Accuracy: {:.2f}%".format(accuracy * 100))
```

```python
#Predict on test
test_prediction = np.argmax(y_pred, axis=1)
#Inverse le transform to get original label back. 
test_prediction = le.inverse_transform(test_prediction)
```

```python
test_prediction
```

```python
#Print overall accuracy
from sklearn import metrics
print ("Accuracy = ", metrics.accuracy_score(test_labels, test_prediction))
```
